## Base Model Regression

```python
# Necessary imports
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import patsy

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
%matplotlib inline
```

```python
music_df.info()
```

```python
# Target vector (y)
y_music_df = music_df['peak_rank']

# Feature matrix (X)
X_music_df = music_df.drop(columns = ['peak_rank','artist_name','song_title','Week_of'])
```

## Base Model Regression with sklearn

```python
# create an empty model
skl_lm = LinearRegression()

# Fit the model to the full dataset
skl_lm.fit(X_music_df, y_music_df)

# Print out the R^2 for the model against the full dataset
sklearn_baseline_r2 = skl_lm.score(X_music_df,y_music_df)
sklearn_baseline_r2
```

## Base Model Regression with Statsmodel

```python
# Create Stats model
sm_lm = smf.ols('peak_rank ~ last_week_rank + weeks_on_chart + danceability + energy + key + loudness + mode + speechiness + acousticness + instrumentalness + liveness + valence + tempo + duration_ms + time_signature + current_rank', data=music_df)

# Fit your model to your training set
sm_fit = sm_lm.fit()

# Print summary statistics of the model's performance
statsmodel_baseline_r2 = sm_fit.summary()
statsmodel_baseline_r2
```

## Resideual Plot of Base Model

```python
plt.figure(figsize=(12,8))
plt.scatter(sm_fit.predict(), sm_fit.resid)
plt.xlabel("predicted target values")
plt.ylabel("residuals")
plt.title("Residual Plot: Baseline Model")
```

```python
# The resideuals are normally distributed meaning I can use a linear regression to model my data
```

## Model Optimization

```python
# imports
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso #ordinary linear regression + w/ ridge regularization
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
```

```python
# hold out 20% of the data for final testing (80 rows of data for testing)
X, X_test, y, y_test = train_test_split(X_music_df, y_music_df, test_size=.2, random_state=10)
```

```python
# split the data into Train/Validate sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=.25, random_state=3)
```

```python
# Linear Regression model
skl_lm = LinearRegression()

#Feature scaling for train, val, and test so that we can run our ridge model on each
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train.values)
X_val_scaled = scaler.transform(X_val.values)
X_test_scaled = scaler.transform(X_test.values)

# Lasso regression model
lm_lasso = Lasso(alpha=1)

# Ridge regression model
lm_ridge = Ridge(alpha=1)

#Feature transform for train, val, and test so that we can run our poly model on each
poly = PolynomialFeatures(degree=2) 

X_train_poly = poly.fit_transform(X_train.values)
X_val_poly = poly.transform(X_val.values)
X_test_poly = poly.transform(X_test.values)

# Degree 2 polynomial regression model
lm_poly = LinearRegression()
```

### Score Models

```python
# Score models

# linear regression
skl_lm.fit(X_train, y_train)
print(f'Linear Regression train R^2: {skl_lm.score(X_train, y_train):.3f}')
print(f'Linear Regression val R^2: {skl_lm.score(X_val, y_val):.3f}')

# linear regression scaled
skl_lm.fit(X_train, y_train)
print(f'Linear Regression on scaled data train R^2: {skl_lm.score(X_train_scaled, y_train):.3f}')
print(f'Linear Regression on scaled data val R^2: {skl_lm.score(X_val_scaled, y_val):.3f}')

# lasso regression
lm_lasso.fit(X_train_scaled, y_train)
print(f'Lasso Regression train R^2: {lm_lasso.score(X_train_scaled, y_train):.3f}')
print(f'Lasso Regression val R^2: {lm_lasso.score(X_val_scaled, y_val):.3f}')

# ridge regression
lm_ridge.fit(X_train_scaled, y_train)
print(f'Ridge Regression train R^2: {lm_ridge.score(X_train_scaled, y_train):.3f}')
print(f'Ridge Regression val R^2: {lm_ridge.score(X_val_scaled, y_val):.3f}')

# degree 2 polynominal regression
lm_poly.fit(X_train_poly, y_train)
print(f'Degree 2 polynomial regression train R^2: {lm_poly.score(X_train_poly, y_train):.3f}')
print(f'Degree 2 polynomial regression val R^2: {lm_poly.score(X_val_poly, y_val):.3f}')
```

```python
# Linear Model: Model is overfitting on the training data by 0.04
# Lasso Model: R2 for training and validation is pretty close
# Ridge Model: Model is similar to Linear model
# Polynomial Regression Model: Model is overfitting on training data
```

```python
# coeficients of lasso regression model 
lm_lasso.coef_
```

```python
# Fit Linear model on (test data) and score model on test data
skl_lm.fit(X,y)
print(f'Linear Regression test R^2: {skl_lm.score(X_test, y_test):.3f}')
```

```python
# Fit Lasso model on (training data) and score model on test data
lm_lasso.fit(X_train_scaled,y_train)
print(f'Linear Regression test R^2: {lm_lasso.score(X_test_scaled, y_test):.3f}')
```
